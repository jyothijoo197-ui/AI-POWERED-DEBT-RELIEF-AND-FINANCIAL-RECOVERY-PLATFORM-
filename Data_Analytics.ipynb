Loops

```python
# Loops : Loops help us to repeat a task multiple times without having to write the same line of code again and again
# -> for: known iterations
# -> while: until a condition is met
```

```python
# for : A for loop is useful when we know exactly how many times we want to repeat something
```

```python
# sum = 1 + 2+ 3+ 4....
```

```python
sum = 0
for i in range(1, 11):
  sum += i

print(sum)
```

```python
# while: A while loop is used when we want to keep repeating something until a certain condition is no longer true
```

```python
sum = 0
i = 0
while(i < 6):
  sum += i
  i += 1

print(sum)
```

Numpy - Numerical Python

```python
import numpy as np
```

```python
sales = np.array([[1200, 1500, 1700], [1000, 1100, 1300]])  # -> 2 dimensional data
```

```python
# row1 -> north
# row2 -> south
# columns -> months(Jan, Feb, March)
```

```python
# North sales:
sales[0]
```

```python
# Sales in South in the month of may:
sales[1][2]
```

```python
# Calculate total sales (both regions, all 3 months):

np.sum(sales)
```

```python
# Average sales of both the regions:
np.mean(sales, axis = 1)
```

```python
# Average sales of each month:
np.mean(sales, axis = 0)
```
